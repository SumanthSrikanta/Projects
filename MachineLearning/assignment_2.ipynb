```python
import pandas as pd
import numpy as np
df = pd.read_csv ('ML_assignment02_dataset_2021.csv')
def del_duplicate(df):
    df=df.drop_duplicates()
    df.to_csv('del_duplicate.csv')
    return df
def fill_values(df):
    df=df.replace(0, np.nan)
    df = df.fillna(df.groupby(df.columns[3]).transform('mean'))
    df=round(df,2)
    df.to_csv('Missing_values.csv')
    return df
def transform_nominal(df):
    df = pd.concat([df, pd.get_dummies(df[df.columns[0]],df.drop(df.columns[0],inplace=True, axis=1))], axis=1)
    df.to_csv('transform_nominal.csv')
    return df
def normalise_minmax(df):
    result=df
    for col_name in df.columns[1:2]:
        max_value = df[col_name].max()
        min_value = df[col_name].min()
        result[col_name] = (df[col_name] - min_value) / (max_value - min_value)
    df=round(df,2)
    df.to_csv('normalise_minmax.csv')
    return df
```

```python
del_duplicate(df)
```

```python
fill_values(df)
```

```python
transform_nominal(df)
```

```python
normalise_minmax(df)
```
